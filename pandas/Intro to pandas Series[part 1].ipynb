# create the series from numpy array

```python
import numpy as np

x = np.array(['a', 'e', 'i', 'o', 'u'])
```

```python
s = pd.Series(x)
```

```python
s
```

```python
type(s)
```

```python
list(s.index)
```

```python
s.values
```

# create pandas series with specified index

```python
pd.Series(['bmw', 'audi', 'mercedes'])
```

```python
pd.Series(['bmw', 'audi', 'mercedes'], index = ['first_car', 'second_car', 'third_car'])
```

```python
s = pd.Series(['bmw', 'audi', 'mercedes'], index = ['first_car', 'second_car', 'third_car'])
```

```python
s
```

```python
s.index
```

```python
s.values
```

# create pandas series using dictionary

```python
d = {
    'a': 40,
    'b': 88,
    'c': 0
}
```

```python
pd.Series(d)
```

# accessing data from pandas series with position index

```python
s = pd.Series(['bmw', 'audi', 'mercedes'], index = ['first_car', 'second_car', 'third_car'])
```

```python
s
```

```python
s[0]
```

```python
s[1]
```

```python
s[2]
```

```python
s[:2].values
```

```python
s['first_car']
```

# giving name to pandas series

```python
s = pd.Series(['bmw', 'audi', 'mercedes'], index = ['first_car', 'second_car', 'third_car'], name = 'cars')
```
