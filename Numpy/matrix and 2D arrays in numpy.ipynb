# Random matrix generation

* Uniform distribution
* Normal distribution

```python
import numpy as np
```

```python
# seed(0)
```

```python
np.random.seed(5) # 
np.random.randint(0,10, (3,2)) # uniform x ∈ to [0, 10)
```

```python
np.random.seed(5) # 
np.random.uniform(1,10, (3,2)) # uniform x ∈ to [0, 10)
```

# Normal distribution

```python
np.random.normal(10, 2, (3,2)) # µ = 10 , σ = 2
```

```python
np.random.randn(3,2)  # µ = 0 , σ = 1
```
