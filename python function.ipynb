````python
def greet(msg = 'Good morning', name): # msg = default argument 
    print(f"Hello, {name} {msg}")
````

# write a program to find factiorial of given number

```
0! = 1
1! = 1
2! = 2 * 1! = 2 * 1
3! = 3 * 2! = 3 * 2 * 1
4! = 4 * 3! = 4 * 3 * 2 * 1
5! = 5 * 4! = 5 * 4 * 3 * 2 * 1
```

````python
def find_factorial(num):
    if num == 0:
        return 1
    fact = 1
    for i in range(1, num + 1):
        fact = fact * i
    return fact
        
````

````python
find_factorial(5)
````

```
0! = 1
1! = 1
2! = 2 * 1! = 2 * 1
3! = 3 * 2! = 3 * 2 * 1
4! = 4 * 3! = 4 * 3 * 2 * 1
5! = 5 * 4! = 5 * 4 * 3 * 2 * 1
```

````python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)
````
