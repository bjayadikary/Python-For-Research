# Barchart

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
ax.bar(fruits, counts)
ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply Distribution')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
cmap = plt.cm.tab10
colors = cmap(np.arange(len(fruits)) % cmap.N)
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
ax.bar(fruits, counts, color = colors)
ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply Distribution')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 8, 1000)


fig, axes = plt.subplots(2, 2, figsize = (10, 8))
ax = axes[0,0]
ax.plot(x, np.sin(x), color = 'g') #row=0, col=0
ax = axes[0,1]
ax.plot(range(1, 100, 10), 'o', color = 'b') #row=0, col=1
ax = axes[1, 0]
ax.plot(x, np.tan(x), color = 'k') #row=1, col=0
ax = axes[1, 1]
ax.plot(x, np.cos(x), color = 'r') #row=1, col=1
fig.show()
```

```python
# set titles
fig, axes = plt.subplots(2, 2, figsize = (10, 8))
ax = axes[0,0]
ax.set_title("The Sin Function")
ax.plot(x, np.sin(x), color = 'g', label = 'sinx') 
ax.legend()
ax.set_ylabel("$sin(x)$")
ax.set_xlabel("$x$")
ax = axes[0,1]
ax.set_title("Regression Points")
ax.plot(range(1, 100, 10), 'o', color = 'b', label = 'regression') 
ax.legend()
ax = axes[1, 0]
# ax.set_title("Tan function")
ax.plot(x, np.tan(x), color = 'k', label = 'tanx') 
ax.legend()
ax = axes[1, 1]
# ax.set_title("Cos function")
ax.plot(x, np.cos(x), color = 'r', label = 'cosx')
ax.legend()
ax.set_ylabel("$cos(x)$")
ax.set_xlabel("$x$")
fig.tight_layout(pad=4.0)
fig.show()
```

```python
_ = np.linspace(-1, 1, 100)
x, y = np.meshgrid(_,_)
z = x**2+x*y
```

A contour plot is a graphical technique for representing a 3-dimensional surface by plotting constant z slices, called contours, on a 2-dimensional format. 

A contour plot is appropriate if you want to see how some value Z changes as a function of two inputs, X and Y:
z = f(x, y). 

```python
plt.style.use(['default'])
plt.contourf(x,y,z)
```

```python
plt.contourf(x,y,z, levels=20)
plt.colorbar(label='Temperature [$^\circ C$]')
plt.xlabel('Horizontal Position [m]')
plt.ylabel('Vertical Position [m]')
```

```python
plt.figure(figsize = (10,6))
cs = plt.contour(x,y,z, levels=20)
plt.clabel(cs, fontsize=8)
```
