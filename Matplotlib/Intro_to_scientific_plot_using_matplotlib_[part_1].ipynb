# Line Plot

```python
x = np.linspace(0,15, 30)
y = np.sin(x) + 0.1 * np.random.randn(len(x))
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, lw = 3) # lw sets line width
```

```python
plt.plot(x, y, lw = 3, color = 'purple') # color sets color
```

```python
plt.plot(x, y, '--',lw = 3, color = 'purple') # -- means dashed lines
```

```python
plt.plot(x, y, 'o--',lw = 3, color = 'purple') # o-- plots points and connects points with -- lines
```

```python
plt.plot(x, y, 'o--',lw = 3, color = 'purple', ms = 8 ) # ms controls marker strength thickness of o
```

```python
# change figure size
plt.figure(figsize = (8,3))
plt.plot(x, y, 'o--',lw = 3, color = 'purple', ms = 8 ) 
```

```python
# Label the Image on x and y axis
plt.figure(figsize = (8,3))
plt.plot(x, y, 'o--',lw = 3, color = 'purple', ms = 10 ) # color sets color
plt.xlabel('Time [s]', fontsize = 20)
plt.ylabel("Voltage [V]", fontsize = 20)
plt.title("Voltage in Electrode")
plt.show()
```

```python
# adding legend to graph
plt.figure(figsize = (8,4))
plt.plot(x, y, 'o--',lw = 3, color = 'purple', ms = 10, label = "Comp 1" ) # label gives legend
plt.xlabel('Time [s]', fontsize = 20)
plt.ylabel("Voltage [V]", fontsize = 20)
plt.title("Voltage in Electrode")
plt.legend(loc = 'lower right', fontsize = 12)
plt.show()
```

# histogram : frequency distribution

```python
res = np.random.randn(1000) * 0.2 + 0.4
```

```python
# plot histogram
```

```python
plt.hist(res)
```

```python
plt.figure(figsize = (5, 3))
plt.hist(res, bins = 30)
plt.show()
```

# density plot : Histogram plot when integrated Area under histogram 1

density: If True, draw and return a probability density: each bin will display the bin's raw count divided by the total number of counts and the bin width `(density = counts / (sum(counts) * np.diff(bins)))`, so that the area under the histogram integrates to 1 `(np.sum(density * np.diff(bins)) == 1)`.

If stacked is also True, the sum of the histograms is normalized to 1.


```python
plt.figure(figsize = (5, 3))
plt.hist(res, bins = 30, density = True)
plt.show()
```

# plot two histogram in one

```python
np.random.seed(40)
res2 = np.random.randn(1000) * 0.2 +  0.3
```

$$\hat{E} - E$$

```python
plt.figure(figsize = (8, 3))
plt.hist(res, bins = 30, density = True, histtype = 'step')
plt.hist(res2, bins = 30, density = True, histtype = 'step')
plt.xlabel("$\hat{E} - E$", fontsize = 20)
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize = (8, 3))
plt.hist(res, bins = 30, density = True, histtype = 'step', label = "Method 1")
plt.hist(res2, bins = 30, density = True, histtype = 'step', label = "Method 2")
plt.xlabel("$\hat{E} - E$", fontsize = 20)
plt.ylabel("Frequency")
plt.legend(loc= 'upper right')
plt.show()
```

```python
plt.figure(figsize = (8, 3))
plt.hist(res, bins = 30, density = True, histtype = 'step', label = "Method 1")
plt.hist(res2, bins = 30, density = True, histtype = 'step', label = "Method 2")
plt.xlabel("$\hat{E} - E$", fontsize = 20)
plt.ylabel("Frequency")
plt.legend(loc= 'upper right')
plt.savefig('trail.png', dpi = 400) # saves the plotted image locally
plt.show()
```

# adding text on Image

```python
np.random.seed(0)
res1 = np.random.randn(1000) * 0.2 +  0.1
```

```python
std = round(np.std(res1), 3)
std
```

```python
text = "$\sigma_a = %.3f$"%std
```

```python
plt.figure(figsize = (8, 3))
plt.hist(res1, bins = 30, density = True, histtype = 'step', label = 'Method 3')
plt.legend(loc = 'upper right')
plt.text(-0.5, 2, text, bbox = dict(facecolor = 'white', edgecolor = 'purple'))
plt.show()
```
